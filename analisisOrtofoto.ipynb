```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
imagen = cv2.imread('data/1.jpg')
plt.imshow(imagen)
```

```python
b,g,r = cv2.split(imagen)

img2 = cv2.merge([r,g,b])
plt.imshow(img2)
```

```python
hsv = cv2.cvtColor(imagen,cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
```

```python
h0 = 33
h1 = 104
s0 = 41
s1 = 125
v0 = 118
v1 = 204

HSVbajo = np.array((h0,s0,v0))
HSValto = np.array((h1,s1,v1))

mask = cv2.inRange(hsv,HSVbajo,HSValto)
plt.imshow(mask)
```

```python
adt = cv2.addWeighted(imagen,0.7,mask,0.3,0)
plt.imshow(adt)
```
